### Making Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

prep_pipe = ColumnTransformer([
    ('Text Processing', processing_text(),['headline', 'short_description']),
    ('Date Processing', processing_date(),['date'])
], remainder= 'passthrough')

pipe = Pipeline([
    ('preprocessing', prep_pipe)
])
pipe.fit_transform(df_copy.drop(columns='category'))
```
