### Making Pipeline

```python
from sklearn.base import BaseEstimator, TransformerMixin
import re
import string

class processing(BaseEstimator,TransformerMixin) :
    def fit(self, x, y=None ) :
        return self

    def healper(self, x):
        x = re.sub(r'[^a-zA-z\s]','',x.lower())
        return ' '.join(x.split())

    def transform(self, x) :
        combined = (x['headlines'].astype(str) + ' ' +
                    x['description'].astype(str) + ' ' +
                    x['content'].astype(str))

        return combined.apply(self.healper).tolist()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression

text_pipe = Pipeline([
    ('Text',processing()),
    ('Vectorizer', TfidfVectorizer(stop_words='english'))
])

prep_pipe = ColumnTransformer([
    ('Processing', text_pipe,['headlines', 'description', 'content'])
], remainder= 'passthrough')

pipe = Pipeline([
    ('preprocessing', prep_pipe),
    ('Classifier', LogisticRegression(max_iter= 1000, class_weight= 'balanced'))
])
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['enc'] = le.fit_transform(df['category'])
```

```python
df['enc'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['category', 'enc']), df['enc'], test_size= 0.33, random_state= 42)
```

```python
x_train
```

```python
pipe.fit(x_train, y_train)
```

```python
pred = pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

acc = accuracy_score(y_true= y_test, y_pred= pred)

acc
```
