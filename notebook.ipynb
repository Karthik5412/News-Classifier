### Making Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import TfidfVectorizer

text_pipe = Pipeline([
    ('Text',processing_text()),
    ('Vectorizer', TfidfVectorizer(stop_words='english'))
])

prep_pipe = ColumnTransformer([
    ('Processing Heading', text_pipe,['headline']),
    ('Processing Description', text_pipe,['short_description']),
    ('Date Processing', processing_date(),['date']),
], remainder= 'passthrough')

pipe = Pipeline([
    ('preprocessing', prep_pipe)
])
pipe.fit_transform(df_copy.drop(columns='category'))
```

```python

prep_pipe = ColumnTransformer([
    ('Text Processing', processing_text,['headline', 'short_description']),
    ('Date Processing', processing_date(),['date']),
], remainder= 'passthrough')

vec_pipe = ColumnTransformer([
    ('head_vec', TfidfVectorizer(stop_words= 'english'), ['headline']),
    ('dec_vec', TfidfVectorizer(stop_words= 'english'), ['short_description'])
], remainder= 'passthrough')

pipe = Pipeline([
    ('preprocessing', prep_pipe),
    ('Vectorization', vec_pipe)
])
pipe.fit_transform(df_copy.drop(columns='category'))
```
